## Read data

```python
import os
 
# Get the list of all files and directories
path = "./Crypto"
dir_list = os.listdir(path)
```

```python
Cryptos = {}
for file in dir_list: 
    Cryptos[file[5:-4]] = pd.read_csv(path + '/'  + file)
```

```python
for key in Cryptos:
    Cryptos[key]['Date'] = pd.to_datetime(Cryptos[key]['Date']).dt.normalize()
    Cryptos[key].set_index('Date', inplace=True)
```

```python
def crypto_data_explorattion(df, crypto_name=''): 
    display(df.describe())
    fix, ax = plt.subplots(ncols=3, figsize=(18, 6))
    ax[0].set_title('Marketcap ' + crypto_name)
    sns.lineplot(data=df, x=df.index, y='Marketcap', ax=ax[0])
    ax[1].set_title('Close '+ crypto_name)
    sns.lineplot(data=df, x=df.index, y='Close', ax=ax[1])
    ax[2].set_title('Volume '+ crypto_name)
    sns.lineplot(data=df, x=df.index, y='Volume', ax=ax[2])
    plt.show()
    
```

```python
for key, data in Cryptos.items(): 
    crypto_data_explorattion(data, key)
```

```python
returns = pd.concat([Cryptos[cypto].Close.pct_change() for cypto in columns],axis=1)
returns.columns = columns
returns
```

```python
returns['Portfolio Return'] = returns.mean(axis=1)
```

```python
sns.lineplot(data=(1 + returns['Portfolio Return']).cumprod())
```
